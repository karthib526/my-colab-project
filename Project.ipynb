```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Split the single column into multiple columns
df_split = df['fixed acidity;"volatile acidity";"citric acid";"residual sugar";"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";"pH";"sulphates";"alcohol";"quality"'].str.split(';', expand=True)

# Define column names
column_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
df_split.columns = column_names

# Convert columns to numeric, coercing errors
for col in df_split.columns:
    df_split[col] = pd.to_numeric(df_split[col], errors='coerce')

# Drop rows with NaN values that resulted from coercion
df_split.dropna(inplace=True)

# Apply StandardScaler
std = StandardScaler()
df_std = pd.DataFrame(std.fit_transform(df_split), columns=df_split.columns)

display(df_std.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode the target label 'quality'
le = LabelEncoder()
df_split['quality_encoded'] = le.fit_transform(df_split['quality'])

display(df_split.head())
```
